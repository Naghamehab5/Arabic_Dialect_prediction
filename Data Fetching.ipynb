```python
import sqlite3
import pandas as pd
```

```python
db_file = "D:/iti2/NLP/dialects_database.db"
```

```python
connect = sqlite3.connect(db_file)

cursor = connect.cursor()

cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")

tables = cursor.fetchall()

print("Tables in the database:")
for table in tables:
    print(table[0])
```

```python
table1='id_text'
table2='id_dialect'

# join the two tables on the 'id' column
query = f"""
SELECT {table1}.id, {table1}.text, {table2}.dialect
FROM {table1}
JOIN {table2} ON {table1}.id = {table2}.id
"""
df = pd.read_sql_query(query, connect)

connect.close()
pd.set_option('display.max_colwidth', 100000)

df.head(15)
```

```python
df.to_csv('D:/iti2/NLP/final project/Fetching Data/dialects_data.csv', index=False)
```
